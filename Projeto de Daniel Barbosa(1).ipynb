```python
# Pegando os dados em zip
dados_prf_2018 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/MaC6cieXSFACNWT/download',
                        encoding = 'latin1', compression = 'zip', sep = ';')
dados_prf_2019 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/kRBUylqz6DyQznN/download',
                        encoding = 'latin1', compression = 'zip', sep = ';')
dados_prf_2020 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/jdDLrQIf33xXSCe/download',
                        encoding = 'latin1', compression = 'zip', sep = ';')
dados_prf_2021 = pd.read_csv('http://arquivos.prf.gov.br/arquivos/index.php/s/n1T3lymvIdDOzzb/download',
                        encoding = 'latin1', compression = 'zip', sep = ';')

# juntando todo os dataframes com concat
juntando_dados = pd.concat([dados_prf_2018,dados_prf_2019 ,dados_prf_2020,dados_prf_2021],
                          keys = ['2018', '2019', '2020', '2021'])
juntando_dados
```

```python
# Quantidades de acidentes em todos os anos separadamente e juntos nos três anos

acidentes_pb_2018 = dados_prf_2018[(dados_prf_2018.uf == 'PB') & (dados_prf_2018.feridos_graves >= 1)]
acidentes_pb_2019 = dados_prf_2019[(dados_prf_2019.uf == 'PB') & (dados_prf_2019.feridos_graves >= 1)]
acidentes_pb_2020 = dados_prf_2020[(dados_prf_2020.uf == 'PB') & (dados_prf_2020.feridos_graves >= 1)] 
acidentes_pb_2021 = dados_prf_2021[(dados_prf_2021.uf == 'PB') & (dados_prf_2021.feridos_graves >= 1)]
acidentes_pb_3anos = juntando_dados[(juntando_dados.uf == 'PB') & (juntando_dados.feridos_graves >= 1)]

print('A Paraiba teve', len(acidentes_pb_2018), ' de acidentes com pessoas gravemente feridas em 2018')
print('A Paraiba teve', len(acidentes_pb_2019), ' de acidentes com pessoas gravemente feridas em 2019')
print('A Paraiba teve', len(acidentes_pb_2020), ' de acidentes com pessoas gravemente feridas em 2020')
print('A Paraiba teve', len(acidentes_pb_2021), ' de acidentes com pessoas gravemente feridas em 2021')
print('Juntando todos os anos a Paraiba teve', len(acidentes_pb_3anos),
      ' de acidentes com pessoas gravemente feridas em 2021')
```

```python
# separando o ano do mes e do dia

mes18 = dados_prf_2018[['ano', 'mes', 'dia']] = dados_prf_2018['data_inversa'].str.split('-', expand=True)
mes18.iloc[:,1:2].rename(columns = {1:'mes'})

# pegando colunas que interressão para a analise

mes_2018b = dados_prf_2018.loc[:,['mes','pessoas','ilesos','feridos_graves']]
ano2018 = mes_2018b.groupby(by='mes').sum()
probalidade_ilesos_18 = (ano2018['ilesos']/ano2018['pessoas'])
probalidade_feridos_18 = (ano2018['feridos_graves']/ano2018['pessoas'])
probalidade_2018 = pd.DataFrame(probalidade_ilesos_18).rename(columns={0:'proporção ilesos em 2018'}).round(4)
probalidade_2018['proporção feridos graves em 2018'] = pd.DataFrame(probalidade_feridos_18).round(4)
probalidade_2018

```

```python
# manipulação de dados para analise acidente durantes e antes da pandemia e porcentagens
antes_da_pandemia = len(pd.concat([dados_prf_2018,dados_prf_2019]))
durante_pandemia = len(pd.concat([dados_prf_2020,dados_prf_2021]))
antes = antes_da_pandemia /2
durante = durante_pandemia /2
diferença = antes - durante

# porcentagem de aumento ou diminição de casos referente ao ano anterior
porcentagem18_19 = (len(dados_prf_2019)-len(dados_prf_2018))/100
porcentagem19_20 = (len(dados_prf_2020)-len(dados_prf_2019))/100
porcentagem20_21 = (len(dados_prf_2021)-len(dados_prf_2020))/100
```

```python
analise_pandemia = pd.DataFrame([antes,durante],
                                index=['Antes da Pandemia','Durante a Pandemia']
                               ).rename(columns={0:'Media de Incidentes'})
analise_pandemia
```

```python
defere = pd.DataFrame([diferença],index=['Diferença']).rename(columns={0:"Diferença entre Periodos de anos"})
defere
```

Com essa analse da diferença de acidentes de antes da pandemia para durante vê que se tem uma Diferença 4339.0 a menos de acidentes 

```python
porcentagem = pd.DataFrame([porcentagem18_19,porcentagem19_20 ,porcentagem20_21],
                           index=['Porcentagem de 2019 referente 2018','Porcentagem de 2020 referente 2019',
                                 'Porcentagem de 2021 referente 2020']).rename(columns={0:'Porcentagem %'})
porcentagem
```
